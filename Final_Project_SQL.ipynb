## **SQL statement untuk menemukan jumlah customer per country**

```python
# Saya tambah mengurut dari abjad terkecil
sql6 = """ SELECT customer_state, COUNT(customer_id) as total_customer
FROM public.oilist_customer
GROUP BY customer_state
ORDER BY customer_state ASC;"""
customer_per_country = pd.read_sql_query(sql6, conn)
customer_per_country
```

## **SQL statement untuk menemukan jumlah total order per country**

```python
# Saya tambah mengurut dari abjad terkecil
sql7 = """ SELECT t2.customer_state, COUNT(order_id) as Total_Order
FROM public.oilist_orders t1
JOIN public.oilist_customer t2
ON t1.customer_id = t2.customer_id
GROUP BY customer_state
ORDER BY customer_state ASC;"""
order_per_country = pd.read_sql_query(sql7, conn)
order_per_country
```

## **SQL statement untuk menemukan berapa banyak jumlah order yang diterima oleh seller dalam interval 1 bulan**

```python
sql8 = """ SELECT t1.seller_id, CAST(EXTRACT(MONTH FROM TO_DATE(t3.order_purchase_timestamp, 'DD/MM/YYYY')) AS INT) as order_month, COUNT(t2.order_id) as Total_Order
FROM public.oilist_sellers t1
JOIN public.oilist_orders_detail t2 ON t1.seller_id = t2.seller_id
JOIN public.oilist_orders t3 ON t2.order_id = t3.order_id
GROUP BY t1.seller_id, order_month
ORDER BY t1.seller_id, order_month ASC ;"""
monthly_total_order = pd.read_sql_query(sql8, conn)
monthly_total_order
```

## **SQL statement untuk menemukan TOP 10 buyer by total_price dari setiap country**

```python
sql9 = """ SELECT t2.customer_id, t2.customer_state, SUM(t3.price) as Total_Price
FROM public.oilist_orders t1
JOIN public.oilist_customer t2 ON t1.customer_id=t2.customer_id
JOIN public.oilist_orders_detail t3 ON t1.order_id=t3.order_id
GROUP BY t2.customer_id, t2.customer_state
ORDER BY total_price DESC
LIMIT 10;"""
top_10_customer = pd.read_sql_query(sql9, conn)
top_10_customer
```

## **SQL statement untuk menemukan latest order_purchase_timestamp dari setiap buyer**

```python
sql10 = """
SELECT customer_id, MAX(TO_TIMESTAMP(order_purchase_timestamp, 'DD/MM/YYYY HH24.MI')) as order_purchase_timestamp
FROM public.oilist_orders
GROUP BY customer_id, order_purchase_timestamp
ORDER BY order_purchase_timestamp DESC;"""
latest_order_customer = pd.read_sql_query(sql10, conn)
latest_order_customer
```

## **SQL statement untuk menemukan jumlah masing-masing order_status dari masing-masing customer**

```python
sql11 = """
SELECT t2.customer_id, order_status, COUNT(order_status) as jumlah
FROM public.oilist_orders as t1
JOIN public.oilist_customer as t2 ON t1.customer_id = t2.customer_id
GROUP BY t2.customer_id, order_status
ORDER BY t2.customer_id ASC;"""
customer_order_status = pd.read_sql_query(sql11, conn)
customer_order_status
```
